```python
# Función para convertir uint8 arrays a boolean arrays
def uint8_array_to_bool(array):
    return array.astype(bool)

# Leer el archivo HDF5
with h5py.File('921.h5', 'r') as hf:
    # Obtener las claves (nombres de los grupos) dentro del archivo HDF5
    group_keys = list(hf.keys())

    # Acceder a los datos de cada grupo dentro del archivo
    for key in group_keys:
        group = hf[key]
        segmentation = uint8_array_to_bool(group['segmentation'][:])
        area = group['area'][()]
        bbox = group['bbox'][:]
        predicted_iou = group['predicted_iou'][()]
        point_coords = group['point_coords'][:]
        stability_score = group['stability_score'][()]  # Añadí esta línea para obtener stability_score
        crop_box = group['crop_box'][:]

        # Ahora puedes trabajar con los datos obtenidos de cada grupo
        print(f"Datos del grupo {key}:")
        print("Segmentation:")
        print(segmentation)
        print("Area:", area)
        print("Bbox:", bbox)
        print("Predicted IoU:", predicted_iou)
        print("Point Coords:", point_coords)
        
        print("Stability Score:", stability_score)
        print("Crop Box:", crop_box)
        print("\n")
```

```python
def eliminar_grupo(file_name, group_name):
    # Abrir el archivo HDF5 en modo lectura y escritura
    with h5py.File(file_name, 'a') as hf:
        # Verificar si el grupo existe antes de intentar eliminarlo
        if group_name in hf:
            del hf[group_name]
            print(f"Grupo '{group_name}' eliminado correctamente.")
        else:
            print(f"El grupo '{group_name}' no existe en el archivo.")

# Nombre del archivo HDF5
file_name = '1002.h5'

# Nombre del grupo que deseas eliminar
grupo_a_eliminar = 'segmentacion_100'

# Llamar a la función para eliminar el grupo
eliminar_grupo(file_name, grupo_a_eliminar)
```

```python
import pandas as pd
import os
import h5py
```

```python
# Especifica la ruta del archivo xlsx
ruta_archivo = 'EliminarMatrices.xlsx'

# Carga el archivo en un DataFrame de pandas
df = pd.read_excel(ruta_archivo)

# Muestra las primeras filas del DataFrame
df.head()
```

```python
# Supongamos que las columnas que mencionaste están en el DataFrame df
# Puedes adaptar el código según tu DataFrame real

# Transforma las columnas Unnamed 2 a Unnamed 9 en enteros y reemplaza NaN con cero
columnas_a_transformar = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8']

df[columnas_a_transformar] = df[columnas_a_transformar].fillna(0).astype(int)
```

```python
# Cargar el DataFrame desde el archivo CSV (reemplaza 'ruta_a_tu_csv' con la ruta correcta)
#df = pd.read_csv('ruta_a_tu_csv')

# Ruta a la carpeta Matrices
carpeta_matrices = 'Matrices/'

# Iterar a través de los archivos .h5 en la carpeta Matrices
for archivo_h5 in os.listdir(carpeta_matrices):
    if archivo_h5.endswith('.h5'):
        ruta_completa = os.path.join(carpeta_matrices, archivo_h5)
        
        # Obtener el nombre del archivo sin la extensión
        nombre_archivo = os.path.splitext(archivo_h5)[0]

        # Buscar coincidencias en el DataFrame usando id_subcarpeta
        coincidencias = df[df['id_subcarpeta'] == int(nombre_archivo)]

        if not coincidencias.empty:
            # Abrir el archivo HDF5
            with h5py.File(ruta_completa, 'r+') as hf:
                # Obtener las claves (nombres de los grupos) dentro del archivo HDF5
                group_keys = list(hf.keys())

                # Iterar a través de los grupos y eliminar los no deseados
                for key in group_keys:
                    if key.startswith('segmentacion_'):
                        segmentacion_id = int(key.split('_')[1])

                        # Verificar si el segmento debe ser eliminado
                        if segmentacion_id not in coincidencias['V1'].values and segmentacion_id not in coincidencias['V2'].values and segmentacion_id not in coincidencias['V3'].values and segmentacion_id not in coincidencias['V4'].values and segmentacion_id not in coincidencias['V5'].values and segmentacion_id not in coincidencias['V6'].values and segmentacion_id not in coincidencias['V7'].values and segmentacion_id not in coincidencias['V8'].values:
                            del hf[key]

                print(f"Grupos no deseados eliminados para {archivo_h5}")

        else:
            print(f"No se encontraron coincidencias para {archivo_h5}")
```
